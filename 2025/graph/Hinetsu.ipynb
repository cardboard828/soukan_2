```python
import numpy as np
import matplotlib.pyplot as plt

def Hinetsu(x):
    return 7/2-(x/np.sinh(x))**2

# Create x values
x1 = np.linspace(0, 5, 500)

# Calculate y values, handling the singularity at x=0
y1 = np.zeros_like(x1)
for i, val in enumerate(x1):
    if np.isclose(val, 0):
        y1[i] = 0
    else:
        y1[i] = Hinetsu(val)

x2 = np.linspace(0, 5, 20)
y2 = np.ones_like(x2) * 2.5

x3 = np.linspace(0, 5, 20)
y3 = np.ones_like(x3) * 3.5

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x1, y1, label='$C_V$')
plt.plot(x2, y2, "--", color='red')
plt.plot(x3, y3, "--", color='red')
# y軸の表示範囲を5/2からに設定
plt.xlim(0, None)
plt.ylim(2.4, 3.6)
plt.xticks([0, 5])
plt.yticks([5/2, 7/2])
plt.tick_params(labelright=False, labeltop=False, bottom=False, left=False, right=False, top=False)
plt.xlabel('$E\mu/(k_B T)$')
plt.ylabel('$C_V/(Nk_B)$')
plt.savefig('Hinetsu.svg')
```

```python
import numpy as np
import matplotlib.pyplot as plt

def langevin_function_1(x):
    """Calculates the Langevin function L(x) = coth(x) - 1/x."""
    return 1 / np.tanh(5*x) - 1 / (5*x)

def langevin_function_2(x):
    """Calculates the Langevin function L(x) = coth(x) - 1/x."""
    return 1 / np.tanh(0.5*x) - 1 / (0.5*x)

# Create x values
x1 = np.linspace(0, 10, 500)
x2 = np.linspace(0, 10, 500)

# Calculate y values, handling the singularity at x=0
y1 = np.zeros_like(x1)
for i, val in enumerate(x1):
    if np.isclose(val, 0):
        y1[i] = 0
    else:
        y1[i] = langevin_function_1(val)

y2 = np.zeros_like(x2)
for i, val in enumerate(x2):
    if np.isclose(val, 0):
        y2[i] = 0
    else:
        y2[i] = langevin_function_2(val)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x1, y1, label='low')
plt.plot(x2, y2, label='high', color='red')
plt.yticks([0, 1])
plt.tick_params(labelbottom=False, labelright=False, labeltop=False, bottom=False, left=False, right=False, top=False)
plt.legend()
plt.xlabel('$E\mu$')
plt.ylabel('$\mathcal{P}_z/(N\mu/V)$')
plt.savefig('2-langevin.svg')
```
